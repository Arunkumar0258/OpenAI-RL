```python
import numpy as np
import gym
import random
```

```python
env = gym.make('FrozenLake-v0')
```

```python
action_space = env.action_space.n
state_space = env.observation_space.n

print(action_space, state_space)
```

```python
qtable = np.zeros((state_space, action_space))
```

```python
num_episodes = 20000
num_steps = 99
alpha = 0.9
gamma = 0.95

epsilon = 1.0
min_epsilon = 0.01
max_epsilon = 1.0
decay_rate = 0.001
rewards = []
```

```python
for episode in range(num_episodes):
    state = env.reset()
    step = 0
    done = False
    total_rewards = 0
    
    for step in range(num_steps):
        tradeoff = random.uniform(0,1)
        
        if tradeoff > epsilon:
            action = np.argmax(qtable[state,:])
        else:
            action = env.action_space.sample()
            
        new_state, reward, done, info = env.step(action)
        
        qtable[state,action] = qtable[state,action] + alpha*(reward + \
                                    gamma*np.max(qtable[new_state,:]) - qtable[state,action])
        
        total_rewards += reward
        state = new_state
    
    epsilon = min_epsilon + (max_epsilon - min_epsilon)*np.exp(-decay_rate*episode)
    rewards.append(total_rewards)
    
print("Score over time: ", sum(rewards)/num_episodes)
print(qtable)
```

```python
env.reset()

total = 0
test_episodes = 5

for episode in range(test_episodes):
    step = 0
    state = env.reset()
    done = False
    print()
    print("Episode", episode)
    
    for step in range(num_steps):
        action = np.argmax(qtable[state,:])
        
        new_state, reward, done, info = env.step(action)
        
        if done:
            env.render()
            print("Reward is: ", reward)
        
        state = new_state
        total += reward

print("Score over time: ", total/test_episodes)
env.close()
```
